```python
import os
import pandas as pd

# Caminho para a pasta principal onde as subpastas estão localizadas
main_folder = 'data'

# Lista para armazenar os dataframes
dataframes = []

# Percorre todas as subpastas na pasta principal
for subfolder in os.listdir(main_folder):
    subfolder_path = os.path.join(main_folder, subfolder)
    if os.path.isdir(subfolder_path):
        # Percorre todos os arquivos na subpasta
        for file in os.listdir(subfolder_path):
            file_path = os.path.join(subfolder_path, file)
            if os.path.isfile(file_path) and file.endswith('.csv'):
                # Lê o arquivo CSV e adiciona à lista de dataframes
                df = pd.read_csv(file_path, sep=';')  # Ajuste o separador se necessário
                dataframes.append(df)

# Concatena todos os dataframes em um único dataframe
merged_df = pd.concat(dataframes, ignore_index=True)

# Salva o dataframe resultante em um arquivo CSV
merged_df.to_csv('merged_data.csv', index=False, sep=';')  # Especifique o separador para salvar

print("Merge concluído com sucesso. O arquivo foi salvo como 'merged_data.csv'.")

```

```python
import pandas as pd
from dateutil import parser

# Função para corrigir valores na coluna nomeEstacao
def corrigir_nome_estacao(df):
    df['nomeEstacao'] = df['nomeEstacao'].replace(['Alto do Cé‚u', 'Alto do Cé\x82u'], 'Alto do Céu')
    return df

# Função para converter coluna datahora para datetime
def converter_para_datetime(df):
    def parse_date(date_str):
        try:
            return parser.parse(date_str)
        except:
            return pd.NaT  # Retorna NaT se houver erro na conversão
    
    df['datahora'] = df['datahora'].apply(parse_date)
    return df

# Carrega o arquivo CSV corrigido com os parâmetros corretos
df = pd.read_csv('merged_data.csv', sep=';')

# Aplica as funções de correção
df = corrigir_nome_estacao(df)
df = converter_para_datetime(df)

```

```python
# Função para converter vírgula para ponto nas colunas numéricas
def converter_virgula_para_ponto(df):
    # Lista de colunas que podem conter números com vírgula
    colunas_numericas = ['latitude', 'longitude', 'valorMedida']  # Substitua com os nomes das suas colunas
    
    for coluna in colunas_numericas:
        if df[coluna].dtype == 'object':  # Verifica se a coluna é do tipo objeto (string)
            df[coluna] = df[coluna].str.replace(',', '.')  # Substitui vírgulas por pontos
            
        elif df[coluna].dtype == 'float64':  # Se já for do tipo float64, não precisa converter
            pass
    
    return df

df = converter_virgula_para_ponto(df)
df
```
